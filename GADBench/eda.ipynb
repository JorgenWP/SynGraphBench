```python
from utils import Dataset
import pandas as pd

# Load a dataset (e.g., 'reddit', 'weibo', or 'amazon')
data = Dataset('tolokers')
graph = data.graph
```

```python
# Extract features and labels as numpy arrays
features = graph.ndata['feature'].numpy()
labels = graph.ndata['label'].numpy()

# Create a DataFrame
df = pd.DataFrame(features)

# Add the label column (1 for anomaly, 0 for normal)
df['label'] = labels

# View the first few rows
print(df.head())
```

```python
# Apply a split (e.g., trial 0, fully supervised)
data.split(semi_supervised=False, trial_id=0)

# Add masks to the DataFrame
df['train_mask'] = graph.ndata['train_mask'].numpy()
df['val_mask'] = graph.ndata['val_mask'].numpy()
df['test_mask'] = graph.ndata['test_mask'].numpy()

# Filter to see only training nodes
training_nodes_table = df[df['train_mask'] == True]
print(training_nodes_table.head())
```

```python
u, v = graph.edges()
edge_df = pd.DataFrame({
    'source_node': u.numpy(),
    'destination_node': v.numpy()
})
print(edge_df.head())
```
